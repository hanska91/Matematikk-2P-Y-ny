# Oppgave 1 - Del 2 Eksempelsett Våren 2023

Synne og Thea prøver å løse oppgaven nedenfor. 

Verdien av en båt har avtatt med 4 % hvert år de siste 8 årene.
I dag er båtens verdi 45 000 kroner.
Hva var båtens verdi for 8 år siden?

Synne sier: «Når verdien har avtatt med 4 %, må vel vekstfaktoren bli 0,96?»
Thea sier: «Kan vi bruke den vekstfaktoren når vi skal finne verdien for 8 år siden? Var ikke verdien større enn 45 000 kroner da?

Kommenter det Synne og Thea sier, og forklar dem hvordan de kan løse oppgaven.

```python
# Enten prosentvis økning eller minking i flere omganger. Ny verdi = Opprinnelig verdi * Vekstfaktor^n hvor n er tiden i antall år
import math

print("Dette programmet regner ut den nye verdien på et tall "
      "som skal øke eller minke med en viss prosent over tid.")

# ta imot informasjon fra brukeren av programmet
svar = "feil"
while svar == "feil":
    svar = input("Dersom tallet skal øke, skriv 'a'. Dersom tallet skal minke, skriv 'm'.")
    if svar != "a" and svar != "m":
        print("Du skrev inn verken 'a' eller 'm'.")
        svar = "feil"

valg = input("Vil du beregne ny verdi (n), gammel verdi (g), vekstfaktor (v) eller tid (t)? ")

if valg == "n":
    tall = float(input("Skriv inn den opprinnelige verdien: "))
    prosent = float(input("Skriv inn prosenten tallet skal endres med: "))
    tid = float(input("Skriv inn tiden i antall år: "))
    
    if svar == "a":
        vekstfaktor = 1 + prosent / 100
    elif svar == "m":
        vekstfaktor = 1 - prosent / 100
    
    ny_verdi = tall * (vekstfaktor ** tid)
    print(f"Den nye verdien etter {tid:.2f} år er {ny_verdi:.2f}.")

elif valg == "g":
    ny_verdi = float(input("Skriv inn den nye verdien: "))
    prosent = float(input("Skriv inn prosenten tallet skal endres med: "))
    tid = float(input("Skriv inn tiden i antall år: "))
    
    if svar == "a":
        vekstfaktor = 1 + prosent / 100
    elif svar == "m":
        vekstfaktor = 1 - prosent / 100
    
    gammel_verdi = ny_verdi / (vekstfaktor ** tid)
    print(f"Den opprinnelige verdien var {gammel_verdi:.2f}.")

elif valg == "v":
    gammel_verdi = float(input("Skriv inn den opprinnelige verdien: "))
    ny_verdi = float(input("Skriv inn den nye verdien: "))
    tid = float(input("Skriv inn tiden i antall år: "))
    
    vekstfaktor = (ny_verdi / gammel_verdi) ** (1 / tid)
    prosent = (vekstfaktor - 1) * 100 if svar == "a" else (1 - vekstfaktor) * 100
    print(f"Vekstfaktoren er {vekstfaktor:.2f}, som tilsvarer en prosentvis endring på {prosent:.2f}%.")

elif valg == "t":
    gammel_verdi = float(input("Skriv inn den opprinnelige verdien: "))
    ny_verdi = float(input("Skriv inn den nye verdien: "))
    prosent = float(input("Skriv inn prosenten tallet skal endres med: "))
    
    if svar == "a":
        vekstfaktor = 1 + prosent / 100
    elif svar == "m":
        vekstfaktor = 1 - prosent / 100
    
    tid = math.log(ny_verdi / gammel_verdi) / math.log(vekstfaktor)
    print(f"Tiden det tar for verdien å endres fra {gammel_verdi:.2f} til {ny_verdi:.2f} er {tid:.2f} år.")
else:
    print("Ugyldig valg.")
```

# Oppgave 3 - Del 2 Eksempelsett Våren 2023

Turistbesøk i nærområdet

Elevene i en klasse har funnet ut hvor mange turister som har besøkt ulike attraksjoner i nærområdet de siste årene. Tabellen nedenfor viser resultatene.

| År | 2015 | 2016 | 2017 | 2018 | 2019 | 2020 | 2021 |
|----|------|------|------|------|------|------|------|
| Totalt antall besøkende | 138 670 | 191 120 | 229 900 | 310 150 | 435 250 | 571 210 | 731 230 |

La x være antall år etter 2015.

a) Bestem et uttrykk på formen: $$ T(x) = a \cdot b^x $$ som er en god modell for totalt antall besøkende de første årene etter 2015.

b) Gi en praktisk tolkning av verdiene du fant for a og b i oppgave a), og vurder modellens gyldighetsområde.

c) Bestem stigningstallet til den rette linjen som går gjennom punktene: $$(1, T(1))$$ og $$(6, T(6))$$
Gi en praktisk tolkning av svaret.

d) Bestem stigningstallene til tangentene til grafen til T i punktene: $$(1, T(1))$$ og $$(6, T(6))$$
Gi en praktisk tolkning av svarene.

```python
# Eksponentialregresjon med dummy data - kode
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, Math

# Datasett
x = np.array([0, 1, 2, 3, 4, 5, 6])                                            # Endre tallene så de passer til din oppgave
y = np.array([138670, 191120, 229900, 310150, 435250, 571210, 731230])         # Endre tallene så de passer til din oppgave

# Utfør eksponentiell regresjon ved hjelp av np.polyfit
log_y = np.log(y)
coeffs = np.polyfit(x, log_y, 1)
a = np.exp(coeffs[1])
b = np.exp(coeffs[0])

# Skriv ut eksponentiell funksjon med LaTeX
latex_str = f"Eksponentialfunksjon: f(x) = {a:.1f} \\cdot {b:.3f}^x"
display(Math(latex_str))

# Skriv ut eksponentiell funksjon uten LaTeX
print(f"Eksponentialfunksjon: f(x) = {a:.1f} * {b:.3f}^x")

# Plot datasettet og eksponentiell funksjon
plt.scatter(x, y, label='Inndata')
plt.plot(x, a * b ** x, label='Eksponentiell funksjon', color="r")
plt.xlabel('Antall år etter 2015')
plt.ylabel('Totalt antall besøkende')
plt.legend()
plt.grid(False)
plt.title("Turistbesøk i nærområdet")
plt.show()

# Beregn vekstfaktor og prosentvis økning/minking
vekstfaktor = b
prosentvis_endring = (vekstfaktor - 1) * 100
print(f"Vekstfaktor b: {vekstfaktor:.3f}")
print(f"Prosentvis endring i hele dataperioden per x-enhet er: {prosentvis_endring:.2f} %")
print(f"Startverdi a er: {a:.1f}")

# Finn y-verdien når x = …
x_val = 60                                                    # Endre denne x verdien for å finne tilhørende y-verdi
y_val = a * (b ** x_val)
print(f"y-verdi for x={x_val}: {y_val:.2f}")

# Finn x-verdien når y = ...
y_val = 500                                                  # Endre denne y verdien for å finne tilhørende x-verdi
x_val = np.log(y_val/a) / np.log(b)
print(f"x-verdi for y={y_val:.2f}: {x_val:.2f}")

# Definer funksjoner
def power_func(x, a, b):
    return a * (b ** x)

def power_func_derivative(x, a, b):
    return a * b ** x * np.log(b)

# Finn stigningstallet mellom to punkter (rett linje)
x1 = 1                                                     # Endre denne x verdien
x2 = 6                                                    # Endre denne x verdien

y1 = power_func(x1, a, b)
y2 = power_func(x2, a, b)

m = (y2 - y1) / (x2 - x1)
print("Stigningstallet mellom punktene ({:.2f}, {:.2f}) og ({:.2f}, {:.2f}) er: {:.1f}".format(x1, y1, x2, y2, m))

# Finn stigningstallet til tangenten i et valgt punkt
x3 = 1                                                # Endre denne x-verdien for å finne stigningstallet til tangenten i punktet
y3 = power_func(x3, a, b)

m_tangent = power_func_derivative(x3, a, b)
print("Stigningstallet til tangenten i punktet ({}, {:.2f}) er: {:.1f}".format(x3, y3, m_tangent))

# Finn stigningstallet til tangenten i et valgt punkt
x4 = 6                                                #  Endre denne x-verdien for å finne stigningstallet til tangenten i punktet
y4 = power_func(x4, a, b)

m_tangent = power_func_derivative(x4, a, b)
print("Stigningstallet til tangenten i punktet ({}, {:.2f}) er: {:.1f}".format(x4, y4, m_tangent))

# Beregn R²-verdien
y_pred = a * b ** x
ss_res = np.sum((y - y_pred) ** 2)
ss_tot = np.sum((y - np.mean(y)) ** 2)
r2 = 1 - (ss_res / ss_tot)
print(f"R²-verdi: {r2:.2f}")
```

# Oppgave 4 - Del 2 Eksempelsett Våren 2023

På en nettside finner Johan teksten i rammen nedenfor.
 
Formelen for omregning mellom celsiusgrader, C, og fahrenheitgrader, F, er

$$F = \frac{9}{5} \cdot C + 32$$

Vurder påstanden om at en temperaturendring på 5 °C tilsvarer en temperaturendring på 41 °F. 

```python
# Omgjøring mellom Celsius, Farenheit og Kelvin
def celsius_to_fahrenheit(tempC):
    return 1.8 * tempC + 32
def celsius_to_kelvin(tempC):
    return tempC + 273.15

def kelvin_to_celsius(tempK):
    return tempK - 273.15

def fahrenheit_to_celsius(tempF):
    return (tempF - 32) / 1.8

def konverter_temperatur():
    enhet = input("Er temperaturen i Celsius (C), Kelvin (K) eller Fahrenheit (F)? ").strip().upper()
    try:
        temperatur = float(input("Hva er temperaturen? "))
    except ValueError:
        print("Ugyldig temperatur. Vennligst oppgi et tall.")
        return

    if enhet == "C":
        tempF = celsius_to_fahrenheit(temperatur)
        tempK = celsius_to_kelvin(temperatur)
        resultat = f"{temperatur} °C er lik {round(tempF, 1)} °F og {round(tempK, 2)} K"
    elif enhet == "K":
        tempC = kelvin_to_celsius(temperatur)
        tempF = celsius_to_fahrenheit(tempC)
        resultat = f"{temperatur} K er lik {round(tempC, 2)} °C og {round(tempF, 1)} °F"
    elif enhet == "F":
        tempC = fahrenheit_to_celsius(temperatur)
        tempK = celsius_to_kelvin(tempC)
        resultat = f"{temperatur} °F er lik {round(tempC, 2)} °C og {round(tempK, 2)} K"
    else:
        resultat = "Ugyldig enhet. Vennligst oppgi C, K eller F."

    print(resultat)

konverter_temperatur()
```

```python
print("Påstanden er feil. En temperaturendring på 5 grader celsius tilsvarer en temperaturendring på 9 grader fahrenheit. De som har lagd nettsiden, har trodd at 0 °F er det samme som 0 °C, slik at forskjellen blir 41 °F. Eller: 5 °C tilsvarer 41 °F, men ikke en endring på 41 °F.")
```

# Oppgave 6 - Del 2 Eksempelsett Våren 2023

Malene er sykepleier. I 2021 hadde hun en inntekt på 490 000 kroner.

Malene ønsker å kjøpe en leilighet. Leiligheten koster 2 500 000 kroner.
På internett finner hun informasjonen nedenfor.

For å få boliglån må du ha betjeningsevne, betalingsvilje, sikkerhet og egenkapital.
Du kan ikke låne mer enn 5 ganger inntekten din. Dette gjelder totalt lån inklusive studielån og forbrukslån.
Du må ha minimum 15 % egenkapital til boligkjøp.

 
Malene har 400 000 kroner i egenkapital og 250 000 kroner i studielån.
Gjør beregninger og finn ut om Malene kan få et stort nok lån til å kunne kjøpe leiligheten hun ønsker seg.

```python
# Malene's inntekt og økonomiske detaljer
inntekt = 490000
leilighet_pris = 2500000
egenkapital = 400000
studielan = 250000

# Beregn maksimalt lånebeløp basert på inntekt (5 ganger inntekten)
maks_lan_belop_basert_pa_inntekt = 5 * inntekt

# Beregn krav til egenkapital (15 % av leilighetens pris)
krav_til_egenkapital = 0.15 * leilighet_pris

# Beregn totalt lån nødvendig (leilighetens pris - egenkapital)
totalt_lan_nodvendig = leilighet_pris - egenkapital

# Beregn maksimalt lån for leiligheten etter fradrag av studielån
maks_lan_for_leilighet_etter_studielan = maks_lan_belop_basert_pa_inntekt - studielan

# Sjekk om Malene kan få et stort nok lån
kan_fa_lan = (totalt_lan_nodvendig <= maks_lan_for_leilighet_etter_studielan and 
              egenkapital >= krav_til_egenkapital)

# Print resultatene med forklaringer
print(f"Malene's inntekt: {inntekt} NOK")
print(f"Leilighetens pris: {leilighet_pris} NOK")
print(f"Malene's egenkapital: {egenkapital} NOK")
print(f"Malene's studielån: {studielan} NOK")
print(f"Maksimalt lånebeløp basert på inntekt (5 ganger inntekten): {maks_lan_belop_basert_pa_inntekt} NOK")
print(f"Krav til egenkapital (15 % av leilighetens pris): {krav_til_egenkapital} NOK")
print(f"Totalt lån nødvendig (leilighetens pris - egenkapital): {totalt_lan_nodvendig} NOK")
print(f"Maksimalt lån for leiligheten etter fradrag av studielån: {maks_lan_for_leilighet_etter_studielan} NOK")

if kan_fa_lan:
    print("Malene kan få et stort nok lån til å kjøpe leiligheten hun ønsker seg.")
    print("Men hun er svært nær den maksimale lånegrensa. Dette kan bety at hun får problemer dersom noe uforutsett skulle skje, for eksempel at renta går mye opp.")
    print("Hvordan banken vurderer betalingsviljen, er ikke godt å si …")
else:
    print("Malene kan ikke få et stort nok lån til å kjøpe leiligheten hun ønsker seg.")
```
