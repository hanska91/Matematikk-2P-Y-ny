```python

from IPython.display import display, HTML
import json
import html as _html

# =========================
# 1) TILPASS HER
# =========================
SIGNATUR = "Christian"  # <- endre om du vil

ELEV_MOTTAKER = "KJ√ÜRE ELEVER"           # f.eks. "KJ√ÜRE VG1 2P-Y"
KOLLEGA_MOTTAKER = "KJ√ÜRE KOLLEGAER"     # f.eks. "KJ√ÜRE TEAMET"

ELEV_TEKST = (
    "JEG H√ÖPER DERE F√ÖR EN MAGISK JULEFERIE ‚Äì MED RO, GODE FOLK RUNDT DERE OG TID TIL √Ö LADE BATTERIENE.<br>"
    "TAKK FOR INNSATSEN I √ÖR! JEG ER STOLT AV DERE OG ALT DERE HAR F√ÖTT TIL.<br><br>"
    "‚≠ê 2026 = NYE MULIGHETER, MESTRING OG MASSE GODE √òYEBLIKK!<br>"
    "TA VARE P√Ö DERE SELV OG HVERANDRE ‚Äì VI SEES ETTER FERIEN!"
)

KOLLEGA_TEKST = (
    "JEG H√ÖPER DERE F√ÖR EN VELFORTJENT PAUSE ‚Äì MED VARME, HVILE OG LITT EKSTRA OVERSKUDD.<br>"
    "TUSEN TAKK FOR SAMARBEIDET I √ÖRET SOM HAR G√ÖTT. JEG SETTER STOR PRIS P√Ö FELLESSKAPET, HUMOREN OG ALT VI F√ÖR TIL SAMMEN.<br><br>"
    "‚≠ê JEG √òNSKER DERE EN STR√ÖLENDE AVSLUTNING P√Ö 2025 OG ET FANTASTISK 2026 ‚Äì MED HELSE, ARBEIDSGLEDE OG GODE DAGER!"
)

# =========================
# 2) TRYGG INNSETTING
# =========================
sig_html = _html.escape(SIGNATUR)
elev_to_html = _html.escape(ELEV_MOTTAKER)
koll_to_html = _html.escape(KOLLEGA_MOTTAKER)

elev_to_js = json.dumps(ELEV_MOTTAKER)
koll_to_js = json.dumps(KOLLEGA_MOTTAKER)
elev_body_js = json.dumps(ELEV_TEKST)
koll_body_js = json.dumps(KOLLEGA_TEKST)

# =========================
# 3) HTML + CSS + JS (INSANE MODE)
# =========================
html_doc = r"""
<div id="nyttar-container" style="max-width: 1020px; margin: 12px auto;">
  <style>
    :root{
      --bg0:#050018; --bg1:#0b1a3a; --bg2:#0a3d2e;
      --gold:#ffd36a; --ice:#b9f3ff; --pink:#ff5fd7; --violet:#b36bff; --lime:#7CFFB2;
      --text:#ffffff; --shadow:rgba(0,0,0,.38);
    }

    .stage{
      position:relative; overflow:hidden;
      border-radius:28px;
      box-shadow:0 22px 55px var(--shadow);
      padding:28px 28px 18px 28px;
      color:var(--text);
      font-family:"Segoe UI", system-ui, -apple-system, Arial, sans-serif;

      /* Animated aurora background */
      background:
        radial-gradient(1200px 650px at 18% 18%, rgba(255,255,255,0.08), transparent 60%),
        radial-gradient(900px 520px at 85% 35%, rgba(255,95,215,0.10), transparent 55%),
        radial-gradient(800px 520px at 35% 95%, rgba(185,243,255,0.08), transparent 55%),
        linear-gradient(135deg, var(--bg0), var(--bg1) 45%, var(--bg2));
    }
    .aurora{
      position:absolute; inset:-50%;
      background: conic-gradient(from 180deg,
        rgba(255,211,106,0.18),
        rgba(185,243,255,0.20),
        rgba(255,95,215,0.16),
        rgba(179,107,255,0.18),
        rgba(124,255,178,0.18),
        rgba(255,211,106,0.18));
      filter: blur(38px);
      opacity: 0.28;
      animation: spin 18s linear infinite;
      pointer-events:none;
      mix-blend-mode: screen;
    }
    @keyframes spin{
      0%{transform: rotate(0deg) scale(1.0);}
      50%{transform: rotate(180deg) scale(1.05);}
      100%{transform: rotate(360deg) scale(1.0);}
    }

    /* Neon glitter title */
    .title{
      font-weight: 1000;
      letter-spacing: 2.1px;
      text-transform: uppercase;
      font-size: clamp(42px, 6.2vw, 82px);
      line-height: 1.03;
      margin: 8px 0 10px 0;
      position: relative;
      display: inline-block;
      text-shadow:
        0 0 12px rgba(185,243,255,0.35),
        0 0 22px rgba(255,95,215,0.30),
        0 12px 34px rgba(0,0,0,0.35);
    }
    .title .stroke{
      -webkit-text-stroke: 2px rgba(255,255,255,0.18);
    }
    .title::after{
      content:"";
      position:absolute; inset:-16px -40px -14px -40px;
      background: linear-gradient(105deg,
        transparent 0%,
        rgba(255,255,255,0.08) 12%,
        rgba(255,211,106,0.55) 22%,
        rgba(185,243,255,0.65) 34%,
        rgba(255,95,215,0.55) 46%,
        rgba(255,255,255,0.10) 58%,
        transparent 78%);
      filter: blur(9px);
      transform: translateX(-70%);
      animation: shimmer 2.8s linear infinite;
      mix-blend-mode: screen;
      pointer-events:none;
      border-radius: 22px;
    }
    @keyframes shimmer{
      0%{ transform: translateX(-70%); opacity:0.85;}
      50%{ opacity:1;}
      100%{ transform: translateX(70%); opacity:0.85;}
    }

    .subtitle{
      display:flex; gap:10px; flex-wrap:wrap; align-items:center;
      font-size: 15px;
      opacity: 0.97;
      margin-bottom: 12px;
      text-transform: uppercase;
      letter-spacing: 0.8px;
      font-weight: 900;
    }
    .chip{
      display:inline-block;
      padding: 7px 12px;
      border-radius: 999px;
      background: rgba(255,255,255,0.10);
      border: 1px solid rgba(255,255,255,0.20);
      box-shadow: 0 10px 24px rgba(0,0,0,0.18);
    }
    .chip.glow{
      background: linear-gradient(90deg, rgba(255,211,106,0.28), rgba(185,243,255,0.22), rgba(255,95,215,0.22));
      border-color: rgba(255,255,255,0.35);
    }

    /* Tabs / buttons */
    .tabs{ display:flex; gap:10px; flex-wrap:wrap; margin: 12px 0 16px 0; }
    .btn{
      border: 1px solid rgba(255,255,255,0.22);
      background: rgba(255,255,255,0.10);
      color: #fff;
      padding: 10px 14px;
      border-radius: 14px;
      cursor: pointer;
      font-weight: 1000;
      letter-spacing: 0.9px;
      text-transform: uppercase;
      transition: transform .12s ease, background .2s ease, border-color .2s ease;
      user-select:none;
      box-shadow: 0 12px 26px rgba(0,0,0,0.18);
    }
    .btn:hover{ transform: translateY(-1px); background: rgba(255,255,255,0.16); border-color: rgba(255,255,255,0.38); }
    .btn.active{
      background: linear-gradient(90deg, rgba(255,211,106,0.35), rgba(185,243,255,0.26), rgba(255,95,215,0.26));
      border-color: rgba(255,255,255,0.50);
    }

    /* Message panel */
    .msg{
      position:relative;
      border-radius: 24px;
      padding: 18px 18px 16px 18px;
      background: rgba(0,0,0,0.18);
      border: 1px solid rgba(255,255,255,0.20);
      backdrop-filter: blur(8px);
    }
    .to{
      font-size: clamp(22px, 2.6vw, 32px);
      font-weight: 1000;
      letter-spacing: 1.3px;
      text-transform: uppercase;
      margin: 0 0 10px 0;
      color: var(--ice);
      text-shadow: 0 0 14px rgba(185,243,255,0.35), 0 10px 28px rgba(0,0,0,0.35);
    }
    .body{
      font-size: 16px;
      line-height: 1.7;
      letter-spacing: 0.35px;
      text-transform: uppercase;
      font-weight: 900;
    }
    .sig{
      margin-top: 14px;
      font-size: 16px;
      font-weight: 1000;
      letter-spacing: 0.8px;
      text-transform: uppercase;
      opacity: 0.98;
    }

    /* Sparkles (CSS) */
    .sparkle{
      position:absolute;
      width: 11px; height: 11px;
      border-radius: 50%;
      background: radial-gradient(circle, rgba(255,255,255,0.98), rgba(255,255,255,0.0) 70%);
      filter: drop-shadow(0 0 10px rgba(255,211,106,0.95));
      opacity: 0;
      animation: twinkle 2.6s ease-in-out infinite;
      pointer-events:none;
    }
    @keyframes twinkle{
      0%,100%{ transform: scale(0.8); opacity:0; }
      50%{ transform: scale(1.75); opacity:1; }
    }

    /* Canvas overlay for fireworks + confetti + snow */
    #fx{
      position:absolute; inset:0;
      width:100%; height:100%;
      pointer-events:none;
      opacity: 0.98;
    }

    .footer{
      display:flex; align-items:center; justify-content:space-between;
      gap:12px; flex-wrap:wrap;
      margin-top: 12px;
      opacity: 0.96;
      font-weight: 850;
      letter-spacing: 0.4px;
      text-transform: uppercase;
    }
    .tiny{ font-size: 12px; opacity: 0.85; text-transform:none; letter-spacing:0.0px; }

    .tog{
      display:inline-flex; align-items:center; gap:10px;
      border:1px solid rgba(255,255,255,0.20);
      background: rgba(255,255,255,0.08);
      padding: 8px 10px;
      border-radius: 999px;
      cursor:pointer;
      user-select:none;
    }
    .dot{
      width: 10px; height: 10px; border-radius: 50%;
      background: var(--lime);
      box-shadow: 0 0 12px rgba(124,255,178,0.75);
    }

    /* Reduced motion */
    @media (prefers-reduced-motion: reduce){
      .title::after, .sparkle, .aurora{ animation:none !important; }
    }
  </style>

  <div class="stage" id="stage">
    <div class="aurora"></div>
    <canvas id="fx"></canvas>

    <!-- CSS sparkles -->
    <div class="sparkle" style="top: 20px; left: 18px; animation-delay: .2s;"></div>
    <div class="sparkle" style="top: 62px; left: 46%; animation-delay: 1.1s;"></div>
    <div class="sparkle" style="top: 34px; right: 18px; animation-delay: .6s;"></div>
    <div class="sparkle" style="bottom: 20px; left: 18px; animation-delay: .9s;"></div>
    <div class="sparkle" style="bottom: 46px; right: 26px; animation-delay: 1.6s;"></div>

    <div class="subtitle">
      <span class="chip glow">2025‚Äì2026</span>
      <span class="chip">‚ú®üéÜüéá GLITTER</span>
      <span class="chip">‚ùÑÔ∏è SN√òFALL</span>
      <span class="chip">üöÄ RAKETTER</span>
      <span class="chip glow" id="countdown">LASTER COUNTDOWN‚Ä¶</span>
    </div>

    <div class="title"><span class="stroke">GOD JUL</span> OG <span class="stroke">GODT NYTT√ÖR</span>!</div>

    <div class="tabs">
      <div class="btn active" id="btn-elev" onclick="showCard('elev')">üéì ELEVER</div>
      <div class="btn" id="btn-koll" onclick="showCard('koll')">üßë‚Äçüè´ KOLLEGAER</div>
      <div class="btn" onclick="megaBurst()">üéá MEGA RAKETT!</div>
      <div class="btn" onclick="snowBoost()">‚ùÑÔ∏è MER SN√ò!</div>
    </div>

    <div class="msg" id="card">
      <div class="to" id="to">__ELEV_TO_HTML__</div>
      <div class="body" id="body">__ELEV_BODY_HTML__</div>
      <div class="sig">‚Äî __SIGNATUR_HTML__</div>
    </div>

    <div class="footer">
      <div class="tiny">Tips: Skjermbilde = perfekt ‚Äúkort‚Äù i Teams. Notebook m√• v√¶re ‚ÄúTrusted‚Äù for JS/Canvas.</div>

      <div style="display:flex; gap:10px; flex-wrap:wrap;">
        <div class="tog" onclick="toggleFX()" title="Skru av/p√• animasjon">
          <div class="dot" id="dotFX"></div>
          <div>ANIMASJON: <span id="fxState">P√Ö</span></div>
        </div>
        <div class="tog" onclick="toggleSnow()" title="Skru av/p√• sn√∏">
          <div class="dot" id="dotSnow"></div>
          <div>SN√ò: <span id="snowState">P√Ö</span></div>
        </div>
      </div>
    </div>
  </div>

  <script>
    // ===== Data (from Python) =====
    const data = {
      elev: { to: __ELEV_TO_JS__, body: __ELEV_BODY_JS__ },
      koll: { to: __KOLL_TO_JS__, body: __KOLL_BODY_JS__ }
    };

    function showCard(which){
      document.getElementById('to').innerHTML = data[which].to;
      document.getElementById('body').innerHTML = data[which].body;
      document.getElementById('btn-elev').classList.toggle('active', which==='elev');
      document.getElementById('btn-koll').classList.toggle('active', which==='koll');
      megaBurst();
    }

    // ===== Countdown to 2026 =====
    const cdEl = document.getElementById('countdown');
    function updateCountdown(){
      const now = new Date();
      const target = new Date(now.getFullYear() + 1, 0, 1, 0, 0, 0); // Jan 1 next year
      const ms = target - now;
      if(ms <= 0){
        cdEl.textContent = "üéâ GODT NYTT√ÖR 2026!!! üéâ";
        return;
      }
      const s = Math.floor(ms/1000);
      const d = Math.floor(s/86400);
      const h = Math.floor((s%86400)/3600);
      const m = Math.floor((s%3600)/60);
      const sec = s%60;
      cdEl.textContent = "COUNTDOWN: " + d + "D " + h + "T " + m + "M " + sec + "S";
    }
    updateCountdown();
    setInterval(updateCountdown, 1000);

    // ===== Canvas Setup =====
    const canvas = document.getElementById('fx');
    const stage  = document.getElementById('stage');
    const ctx    = canvas.getContext('2d');

    let W=0, H=0, DPR=1;

    function resize(){
      DPR = window.devicePixelRatio || 1;
      const rect = stage.getBoundingClientRect();
      W = Math.max(1, Math.floor(rect.width));
      H = Math.max(1, Math.floor(rect.height));
      canvas.width  = Math.floor(W * DPR);
      canvas.height = Math.floor(H * DPR);
      canvas.style.width = W + "px";
      canvas.style.height = H + "px";
      ctx.setTransform(DPR, 0, 0, DPR, 0, 0);
    }
    resize();
    window.addEventListener('resize', resize);

    function rand(a,b){ return a + Math.random()*(b-a); }
    function clamp(x,a,b){ return Math.max(a, Math.min(b,x)); }

    // ===== Fireworks + Confetti Particles =====
    let particles = [];
    function addParticle(x,y,vx,vy,life,size,hue,kind){
      particles.push({x,y,vx,vy,life,ttl:life,size,hue,kind,rot:rand(0,6.28),vr:rand(-0.22,0.22)});
    }

    function explode(x,y,power=1.0){
      // sparks
      const count = Math.floor(110 * power);
      for(let i=0;i<count;i++){
        const ang = (i/count)*Math.PI*2;
        const sp  = rand(0.9, 4.8)*power;
        const vx  = Math.cos(ang)*sp + rand(-0.6,0.6);
        const vy  = Math.sin(ang)*sp + rand(-0.6,0.6);
        const life= rand(40, 95);
        const size= rand(1.0, 2.8)*(power*0.95);
        const hue = rand(0, 360);
        addParticle(x,y,vx,vy,life,size,hue,"spark");
      }
      // glitter confetti
      const conf = Math.floor(42 * power);
      for(let j=0;j<conf;j++){
        const vx  = rand(-3.2,3.2)*power;
        const vy  = rand(-5.5,-1.4)*power;
        const life= rand(70, 145);
        const size= rand(3.8, 7.6);
        const hues= [45, 185, 305, 270, 120, 15];
        const hue = hues[Math.floor(rand(0,hues.length))];
        addParticle(x,y,vx,vy,life,size,hue,"confetti");
      }
    }

    function burst(){
      const x = rand(W*0.12, W*0.88);
      const y = rand(H*0.16, H*0.56);
      explode(x,y, rand(0.85, 1.35));
    }
    function megaBurst(){
      for(let i=0;i<3;i++){
        setTimeout(()=> explode(rand(W*0.18,W*0.82), rand(H*0.18,H*0.52), rand(1.2, 1.9)), i*120);
      }
    }

    // ===== Snow System =====
    let snowflakes = [];
    let snowOn = true;
    let snowIntensity = 1.0; // can boost
    function makeSnowflake(){
      const r = rand(1.2, 3.8);
      return {
        x: rand(0, W),
        y: rand(-H*0.2, H),
        r: r,
        vy: rand(0.55, 1.65) + r*0.15,
        vx: rand(-0.35, 0.35),
        drift: rand(0.002, 0.012),
        phase: rand(0, Math.PI*2),
        alpha: rand(0.35, 0.95),
      };
    }
    function seedSnow(n){
      for(let i=0;i<n;i++) snowflakes.push(makeSnowflake());
    }
    function snowBoost(){
      snowIntensity = clamp(snowIntensity + 0.35, 1.0, 3.0);
      // add more flakes immediately
      seedSnow(Math.floor(60 * snowIntensity));
    }

    // initial snow
    seedSnow(220);

    // ===== Toggles =====
    let running = true;

    function toggleFX(){
      running = !running;
      document.getElementById('fxState').innerText = running ? "P√Ö" : "AV";
      const dot = document.getElementById('dotFX');
      dot.style.background = running ? "var(--lime)" : "#ff6b6b";
      dot.style.boxShadow = running ? "0 0 12px rgba(124,255,178,0.75)" : "0 0 12px rgba(255,107,107,0.55)";
      if(running) megaBurst();
    }
    function toggleSnow(){
      snowOn = !snowOn;
      document.getElementById('snowState').innerText = snowOn ? "P√Ö" : "AV";
      const dot = document.getElementById('dotSnow');
      dot.style.background = snowOn ? "var(--ice)" : "#ff6b6b";
      dot.style.boxShadow = snowOn ? "0 0 12px rgba(185,243,255,0.70)" : "0 0 12px rgba(255,107,107,0.55)";
    }

    // ===== Render Loop =====
    // "motion blur" clear
    function fadeBackground(){
      ctx.fillStyle = "rgba(0,0,0,0.16)";
      ctx.fillRect(0,0,W,H);
    }

    function drawSnow(){
      if(!snowOn) return;

      // keep flake count reasonable
      const targetCount = Math.floor(220 * snowIntensity);
      while(snowflakes.length < targetCount) snowflakes.push(makeSnowflake());
      while(snowflakes.length > targetCount) snowflakes.pop();

      for(let i=0;i<snowflakes.length;i++){
        const f = snowflakes[i];
        f.phase += f.drift;
        f.x += f.vx + Math.sin(f.phase) * 0.35;
        f.y += f.vy;

        // wrap around
        if(f.y > H + 10){
          f.y = rand(-40, -10);
          f.x = rand(0, W);
        }
        if(f.x < -10) f.x = W + 10;
        if(f.x > W + 10) f.x = -10;

        // draw flake with glow
        ctx.save();
        ctx.globalAlpha = f.alpha;
        ctx.fillStyle = "rgba(255,255,255,0.95)";
        ctx.shadowBlur = 10;
        ctx.shadowColor = "rgba(185,243,255,0.55)";
        ctx.beginPath();
        ctx.arc(f.x, f.y, f.r, 0, Math.PI*2);
        ctx.fill();
        ctx.restore();
      }
    }

    function drawParticles(){
      // update + draw fireworks/confetti
      for(let i=particles.length-1;i>=0;i--){
        const p = particles[i];
        p.life -= 1;
        if(p.life <= 0){ particles.splice(i,1); continue; }

        // physics
        p.vy += (p.kind==="confetti") ? 0.055 : 0.045;
        p.vx *= 0.992; p.vy *= 0.992;
        p.x += p.vx; p.y += p.vy;
        p.rot += p.vr;

        const t = p.life / p.ttl;
        const alpha = Math.min(1, t*t*1.15);

        const color = "hsla(" + p.hue + ", 95%, 65%, " + alpha + ")";
        ctx.save();
        ctx.translate(p.x, p.y);
        ctx.rotate(p.rot);

        if(p.kind === "spark"){
          ctx.fillStyle = color;
          ctx.shadowBlur = 18;
          ctx.shadowColor = color;
          ctx.beginPath();
          ctx.arc(0,0,p.size,0,Math.PI*2);
          ctx.fill();
        } else {
          ctx.fillStyle = color;
          ctx.shadowBlur = 16;
          ctx.shadowColor = color;
          ctx.fillRect(-p.size*0.65, -p.size*0.35, p.size*1.3, p.size*0.7);
        }
        ctx.restore();
      }
    }

    function draw(){
      // always resize-safe (in case notebook layout changes)
      // (lightweight - only checks if changed)
      const rect = stage.getBoundingClientRect();
      const newW = Math.floor(rect.width), newH = Math.floor(rect.height);
      if(newW !== W || newH !== H) resize();

      // clear with fade for trail effect
      fadeBackground();

      // snow can run even if fireworks paused (optional), but we keep it independent
      drawSnow();

      if(running){
        drawParticles();
        // auto fireworks sometimes
        if(Math.random() < 0.060 && particles.length < 1100) burst();
      }

      requestAnimationFrame(draw);
    }

    // kickstart
    for(let i=0;i<3;i++) setTimeout(burst, 220 + i*240);
    setTimeout(megaBurst, 900);
    draw();
  </script>
</div>
"""

# Sett inn tekst/verdier
html_doc = (html_doc
    .replace("__SIGNATUR_HTML__", sig_html)
    .replace("__ELEV_TO_HTML__", elev_to_html)
    .replace("__ELEV_BODY_HTML__", ELEV_TEKST)  # Beholder <br>
    .replace("__ELEV_TO_JS__", elev_to_js)
    .replace("__KOLL_TO_JS__", koll_to_js)
    .replace("__ELEV_BODY_JS__", elev_body_js)
    .replace("__KOLL_BODY_JS__", koll_body_js)
)

display(HTML(html_doc))
```
