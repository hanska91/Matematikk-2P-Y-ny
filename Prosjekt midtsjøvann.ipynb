```python
!pip install ttkbootstrap
import locale
import tkinter as tk
from tkinter import messagebox, ttk
import random

# Lokalitet (for å unngå feil med ttkbootstrap)
try:
    locale.setlocale(locale.LC_ALL, 'en_US.UTF-8')
except locale.Error:
    locale.setlocale(locale.LC_ALL, '')

from ttkbootstrap import Style
locale.setlocale(locale.LC_ALL, '')

# Opprinnelige quiz-data (lagres separat for reset)
original_quiz_data = [
  {
    "question": "Hva er en av hovedårsakene til at biologisk mangfold er viktig?",
    "options": ["A. Det gir oss tilgang til rent vann", "B. Det reduserer behovet for fotosyntese", "C. Det øker mengden nitrogen i lufta", "D. Det gjør at vi slipper å dyrke mat"],
    "answer": "A",
    "explanation": "Biologisk mangfold bidrar til økosystemtjenester som rent vann, mat og klimaregulering."
  },
  {
    "question": "Hva er anbefalt pH-verdi for drikkevann?",
    "options": ["A. 4,0–5,0", "B. 5,5–6,0", "C. 6,5–8,5", "D. 8,5–9,5"],
    "answer": "C",
    "explanation": "Anbefalt pH for drikkevann er mellom 6,5 og 8,5 for å unngå korrosjon og helseskader."
  },
  {
    "question": "Hva er en miljøgift?",
    "options": ["A. Et stoff som alltid er naturlig", "B. Et stoff som brytes raskt ned i naturen", "C. Et stoff som akkumuleres i økosystemer og kan skade helse og miljø", "D. Et stoff som kun finnes i planter"],
    "answer": "C",
    "explanation": "Miljøgifter brytes sakte ned og hoper seg opp i næringskjeder, noe som kan skade både dyr og mennesker."
  },
  {
    "question": "Hva er øvre grense for ammonium i drikkevann?",
    "options": ["A. 0,05 mg/L", "B. 0,1 mg/L", "C. 0,5 mg/L", "D. 5 mg/L"],
    "answer": "C",
    "explanation": "Øvre grense for ammonium i drikkevann og fiskevann er 0,5 mg/L."
  },
  {
    "question": "Hva er en viktig indikator på at vann er forurenset av avføring?",
    "options": ["A. Høyt nitratinnhold", "B. Høyt fosfatinnhold", "C. Høyt nitrittinnhold", "D. Lav hardhet"],
    "answer": "C",
    "explanation": "Nitritt er en indikator på forurensning fra avføring og er svært giftig for fisk og skadelig for mennesker."
  },
  {
    "question": "Hva er en fremmedart?",
    "options": ["A. En art som er naturlig hjemmehørende i Norge", "B. En art som er innført til et område den ikke hører hjemme", "C. En art som er utrydningstruet", "D. En art som lever i havet"],
    "answer": "B",
    "explanation": "Fremmede arter er innført til områder der de ikke naturlig hører hjemme, og kan true lokale arter."
  },
  {
    "question": "Hva er hovedårsaken til at nitrat tilsettes i gjødsel?",
    "options": ["A. Det øker vannets pH", "B. Det fremmer plantevekst", "C. Det reduserer fosfatnivået", "D. Det hindrer algevekst"],
    "answer": "B",
    "explanation": "Nitrat fremmer plantevekst og brukes derfor i gjødsel."
  },
  {
    "question": "Hva er anbefalt øvre grense for nitritt i drikkevann?",
    "options": ["A. 0,01 mg/L", "B. 0,03 mg/L", "C. 0,1 mg/L", "D. 1 mg/L"],
    "answer": "C",
    "explanation": "Anbefalt øvre grense for nitritt i drikkevann er 0,1 mg/L."
  },
  {
    "question": "Hva er en produsent i et økosystem?",
    "options": ["A. Et dyr som spiser planter", "B. En organisme som lager egen næring gjennom fotosyntese", "C. En organisme som bryter ned dødt materiale", "D. En organisme som lever av andre rovdyr"],
    "answer": "B",
    "explanation": "Produsenter er planter og alger som lager egen næring gjennom fotosyntese."
  },
  {
    "question": "Hva er en viktig funksjon til myrer?",
    "options": ["A. De øker nitrogeninnholdet i lufta", "B. De lagrer store mengder karbon", "C. De produserer fosfat", "D. De reduserer pH i havet"],
    "answer": "B",
    "explanation": "Myrer lagrer store mengder karbon og bidrar til å motvirke klimaendringer."
  },
  {
    "question": "Hva skjer når vann har høyt innhold av næringsstoffer som nitrat og fosfat?",
    "options": ["A. Vannet blir renere", "B. Det oppstår eutrofiering med stor algevekst", "C. pH stiger til over 9", "D. Vannet blir hardt"],
    "answer": "B",
    "explanation": "Høyt innhold av næringsstoffer fører til eutrofiering, som gir stor algevekst og oksygenmangel."
  },
  {
    "question": "Hva er en konsekvens av å drenere myr?",
    "options": ["A. Økt karbonlagring", "B. Redusert utslipp av CO2", "C. Tap av biologisk mangfold og frigjøring av karbon", "D. Økt vannkvalitet"],
    "answer": "C",
    "explanation": "Drenering av myr frigjør karbon og reduserer biologisk mangfold."
  },
  {
    "question": "Hva er en miljøgift som tidligere ble brukt som insektmiddel?",
    "options": ["A. PCB", "B. DDT", "C. PFAS", "D. Bly"],
    "answer": "B",
    "explanation": "DDT ble brukt som insektmiddel, men er nå forbudt på grunn av miljøskader."
  },
  {
    "question": "Hva er hardhet i vann et mål på?",
    "options": ["A. Mengden oksygen i vannet", "B. Mengden kalsiumsalter i vannet", "C. Mengden ammonium i vannet", "D. Mengden CO2 i vannet"],
    "answer": "B",
    "explanation": "Hardhet i vann bestemmes av mengden kalsiumsalter som er løst i vannet."
  },
  {
    "question": "Hva er en viktig grunn til å bevare biologisk mangfold?",
    "options": ["A. Det gjør vannet hardere", "B. Det gir oss medisiner og mat", "C. Det øker nitratnivået i vann", "D. Det reduserer karbonlagring"],
    "answer": "B",
    "explanation": "Biologisk mangfold gir oss ressurser som mat, medisiner og råmaterialer."
  },
  {
    "question": "Hva er en indikatorart for rent vann?",
    "options": ["A. Røde fåbørstemark", "B. Fjærmygglarver", "C. Steinfluer", "D. Rottehaler"],
    "answer": "C",
    "explanation": "Steinfluer er indikatorarter for rent vann."
  },
  {
    "question": "Hva er en viktig forskjell mellom gift og miljøgift?",
    "options": ["A. Miljøgifter er alltid naturlige", "B. Miljøgifter brytes raskt ned", "C. Miljøgifter akkumuleres i økosystemer", "D. Gifter finnes bare i planter"],
    "answer": "C",
    "explanation": "Miljøgifter akkumuleres i økosystemer og kan skade både dyr og mennesker over tid."
  },
  {
    "question": "Hva er en konsekvens av innføring av fremmede arter?",
    "options": ["A. Økt biologisk mangfold", "B. Redusert konkurranse mellom arter", "C. Utrydding av lokale arter", "D. Økt karbonlagring"],
    "answer": "C",
    "explanation": "Fremmede arter kan konkurrere ut lokale arter og redusere biologisk mangfold."
  },
  {
    "question": "Hva er anbefalt øvre grense for nitrat i drikkevann?",
    "options": ["A. 10 mg/L", "B. 25 mg/L", "C. 50 mg/L", "D. 90 mg/L"],
    "answer": "B",
    "explanation": "Anbefalt øvre grense for nitrat i drikkevann er 25 mg/L."
  },
  {
    "question": "Hva er en viktig årsak til at fosfat havner i vann?",
    "options": ["A. Luftforurensning", "B. Vaskemidler og matvarer", "C. Nitrogenfiksering", "D. Fotosyntese"],
    "answer": "B",
    "explanation": "Fosfatforurensning kommer hovedsakelig fra vaskemidler og matvarer."
  },
  {
    "question": "Hva er en viktig grunn til at pH i vann kan bli lav?",
    "options": ["A. Høyt oksygeninnhold", "B. Høyt CO2-innhold", "C. Høyt nitratinnhold", "D. Høyt fosfatinnhold"],
    "answer": "B",
    "explanation": "Høyt CO2-innhold senker pH i vannet."
  },
  {
    "question": "Hva er en viktig effekt av PFAS?",
    "options": ["A. Brytes raskt ned i naturen", "B. Akkumuleres i næringskjeden og kan skade helse", "C. Øker oksygennivået i vann", "D. Reduserer pH i jord"],
    "answer": "B",
    "explanation": "PFAS brytes svært sakte ned og akkumuleres i næringskjeden, noe som kan skade helse."
  },
  {
    "question": "Hva er en viktig grunn til at torv brukes som hagejord?",
    "options": ["A. Den er næringsrik", "B. Den holder godt på fuktighet", "C. Den øker pH i jorda", "D. Den inneholder mye kalsium"],
    "answer": "B",
    "explanation": "Torv holder godt på fuktighet, derfor brukes den i hagejord."
  },
  {
    "question": "Hva er en viktig konsekvens av eutrofiering?",
    "options": ["A. Økt oksygeninnhold i vannet", "B. Redusert plantevekst", "C. Oksygenmangel og fiskedød", "D. Økt pH i vannet"],
    "answer": "C",
    "explanation": "Eutrofiering fører til oksygenmangel når døde planter brytes ned, noe som kan gi fiskedød."
  },
  {
    "question": "Hva er en viktig grunn til at vi fortsatt dyrker mat i Norge?",
    "options": ["A. For å redusere biologisk mangfold", "B. For å sikre selvforsyning og beredskap", "C. For å øke importen av matvarer", "D. For å redusere karbonlagring"],
    "answer": "B",
    "explanation": "Matproduksjon i Norge er viktig for selvforsyning og beredskap."
  },
  {
    "question": "Hva er en viktig forskjell mellom næringskjede og næringsnett?",
    "options": ["A. Næringskjede viser flere arter, næringsnett viser én", "B. Næringsnett viser flere sammenkoblede kjeder", "C. Næringskjede inkluderer nedbrytere, næringsnett gjør ikke", "D. Næringsnett gjelder bare planter"],
    "answer": "B",
    "explanation": "Næringsnett viser flere sammenkoblede næringskjeder i et økosystem."
  },
  {
    "question": "Hva er en viktig grunn til at bly ble forbudt i bensin?",
    "options": ["A. Det økte drivstofforbruket", "B. Det forårsaket alvorlige helseskader", "C. Det reduserte CO2-utslipp", "D. Det økte pH i vann"],
    "answer": "B",
    "explanation": "Bly i bensin ble forbudt fordi det forårsaket alvorlige helseskader."
  },
  {
    "question": "Hva er en viktig grunn til at biologisk mangfold gir økosystemer stabilitet?",
    "options": ["A. Flere arter gjør økosystemet mer motstandsdyktig mot forstyrrelser", "B. Det reduserer karbonlagring", "C. Det øker utslipp av klimagasser", "D. Det gjør vannet hardere"],
    "answer": "A",
    "explanation": "Høyt biologisk mangfold gjør økosystemer mer motstandsdyktige mot sykdommer og klimaendringer."
  },
  {
    "question": "Hva var en vanlig feilkilde ved nitrittmålingene under ekskursjonen?",
    "options": ["A. For høy temperatur i vannet", "B. Reagensene fungerte ikke som de skulle", "C. Feil bruk av pH-papir", "D. For lite vannprøve"],
    "answer": "B",
    "explanation": "Flere grupper fikk blank løsning i stedet for gul ved 0 mg/L, trolig på grunn av problemer med reagensene."
  },
  {
    "question": "Hvorfor kan det være forskjell mellom pH målt med kjemisk test og pH-papir?",
    "options": ["A. Fordi pH-papir alltid viser høyere verdi", "B. Fordi kjemiske tester er mindre nøyaktige", "C. Fordi pH-papir er mer følsomt for fargeforstyrrelser", "D. Fordi vannets hardhet påvirker pH-papir"],
    "answer": "C",
    "explanation": "pH-papir kan gi unøyaktige resultater hvis vannet har farge eller partikler som forstyrrer avlesningen."
  },
  {
    "question": "Hvilken parameter i klassifiseringsskjemaet sier noe om begroing på bunnen?",
    "options": ["A. Lukt", "B. Skumdannelse", "C. Bunnens og begroingens utseende", "D. Bunndyr og fisk"],
    "answer": "C",
    "explanation": "Bunnens og begroingens utseende vurderes for å bestemme forurensningsgrad."
  },
  {
    "question": "Hva kjennetegner vann i kvalitetsklasse 1 (lite forurenset) når det gjelder lukt?",
    "options": ["A. Råtten, søt-emmen lukt", "B. Aromatisk algeaktig lukt", "C. Ingen merkbar lukt", "D. Svak søt-emmen lukt"],
    "answer": "C",
    "explanation": "Lite forurenset vann har ingen merkbar lukt, eventuelt litt algeaktig."
  },
  {
    "question": "Når ble PCB forbudt i Norge?",
    "options": ["A. 1979", "B. 1989", "C. 1999", "D. 2009"],
    "answer": "A",
    "explanation": "PCB ble forbudt i Norge i 1979 på grunn av alvorlige miljø- og helseskader."
  },
  {
    "question": "Hva er hovedmålet med den norske prioritetslisten for kjemikalier?",
    "options": ["A. Å redusere CO2-utslipp", "B. Å fase ut bruk og utslipp av farlige stoffer", "C. Å øke produksjonen av kjemikalier", "D. Å forby alle syntetiske stoffer"],
    "answer": "B",
    "explanation": "Målet er å fase ut bruk og utslipp av prioriterte farlige stoffer i Norge."
  },
  {
    "question": "Hva skjer med karbonlagringen i myr når den dreneres?",
    "options": ["A. Karbonlagringen øker", "B. Karbon frigjøres som CO2", "C. Karbon omdannes til fosfat", "D. Karbon blir bundet i torv"],
    "answer": "B",
    "explanation": "Når myr dreneres, frigjøres karbon som CO2, noe som bidrar til klimaendringer."
  },
  {
    "question": "Hva er en viktig grunn til at torvuttak er problematisk for miljøet?",
    "options": ["A. Det øker pH i vannet", "B. Det reduserer karbonlagring og ødelegger leveområder", "C. Det gir mer næringsstoffer til myra", "D. Det øker biologisk mangfold"],
    "answer": "B",
    "explanation": "Torvuttak reduserer karbonlagring og ødelegger leveområder for mange arter."
  },
  {
    "question": "Hva er en viktig grunn til at internasjonalt samarbeid er nødvendig for å redusere miljøgifter?",
    "options": ["A. Fordi miljøgifter brytes raskt ned", "B. Fordi miljøgifter spres over landegrenser via luft og hav", "C. Fordi miljøgifter bare finnes i Norge", "D. Fordi miljøgifter ikke påvirker mennesker"],
    "answer": "B",
    "explanation": "Miljøgifter kan spres over store avstander via luft- og havstrømmer, derfor kreves internasjonalt samarbeid."
  },
  {
    "question": "Hva er en viktig forskjell mellom gift og miljøgift?",
    "options": ["A. Miljøgifter brytes raskt ned", "B. Miljøgifter akkumuleres i næringskjeden", "C. Gifter finnes bare i planter", "D. Miljøgifter er alltid naturlige"],
    "answer": "B",
    "explanation": "Miljøgifter akkumuleres i næringskjeden og kan skade både dyr og mennesker over tid."
  }


]

# Initialiser GUI
root = tk.Tk()
root.title("Quiz App")
root.geometry("600x550")
style = Style(theme="flatly")
style.configure("TLabel", font=("Helvetica", 16))
style.configure("TButton", font=("Helvetica", 14))

# Globale variabler
score = 0
current_question = 0
choice_btns = []
quiz_data = []
user_answers = []

# Widgets
qs_label = ttk.Label(root, anchor="center", wraplength=500, padding=10)
qs_label.pack(pady=10)

choices_frame = ttk.Frame(root)
choices_frame.pack(pady=10, fill=tk.X)

feedback_label = ttk.Label(root, anchor="center", padding=10, wraplength=500, justify="left")
feedback_label.pack(pady=10)

score_label = ttk.Label(root, text="", anchor="center", padding=10)
score_label.pack(pady=10)

nav_frame = ttk.Frame(root)
nav_frame.pack(pady=10)

prev_btn = ttk.Button(nav_frame, text="Tilbake", command=lambda: previous_question(), state="disabled")
prev_btn.pack(side="left", padx=10)

next_btn = ttk.Button(nav_frame, text="Neste", command=lambda: next_question(), state="disabled")
next_btn.pack(side="left", padx=10)

retry_btn = ttk.Button(root, text="Prøv igjen", command=lambda: restart_quiz(), state="disabled")
retry_btn.pack(pady=10)

progress = ttk.Progressbar(root, orient="horizontal", length=400, mode="determinate")
progress.pack(pady=10)

# --- Funksjoner ---

def restart_quiz():
    global score, current_question, choice_btns, quiz_data, user_answers
    score = 0
    current_question = 0
    quiz_data = original_quiz_data.copy()
    random.shuffle(quiz_data)
    user_answers = [None] * len(quiz_data)
    progress.config(maximum=len(quiz_data), value=0)
    retry_btn.config(state="disabled")
    next_btn.config(state="disabled")
    prev_btn.config(state="disabled")
    show_question()

def show_question():
    global choice_btns
    for btn in choice_btns:
        btn.destroy()
    choice_btns = []

    question = quiz_data[current_question]
    qs_label.config(text=question["question"])
    options = question["options"]

    selected_index = user_answers[current_question]

    for i, option in enumerate(options):
        button = ttk.Button(
            choices_frame,
            text=option,
            command=lambda i=i: check_answer(i),
            style="TButton"
        )
        button.pack(pady=5, fill=tk.X)
        choice_btns.append(button)

    feedback_label.config(text="")
    score_label.config(text="Score: {}/{}".format(score, len(quiz_data)))
    prev_btn.config(state="normal" if current_question > 0 else "disabled")
    next_btn.config(state="disabled")
    retry_btn.config(state="disabled")

    if selected_index is not None:
        apply_feedback_colors(selected_index)
        next_btn.config(state="normal")

def check_answer(choice_index):
    global score
    question = quiz_data[current_question]
    selected_letter = chr(65 + choice_index)
    correct_letter = question["answer"]
    correct_index = ord(correct_letter) - 65
    previous = user_answers[current_question]

    # Juster score hvis bytter svar
    if previous is not None and chr(65 + previous) == correct_letter:
        score -= 1

    user_answers[current_question] = choice_index

    if selected_letter == correct_letter:
        score += 1
        feedback_label.config(text="Riktig!\n" + question['explanation'], foreground="green")
    else:
        feedback_label.config(
    text=f"Galt! Det riktige svaret er {correct_letter}.\n{question['explanation']}",
    foreground="red"
        )

    score_label.config(text="Score: {}/{}".format(score, len(quiz_data)))
    progress['value'] = current_question + 1

    apply_feedback_colors(choice_index)
    next_btn.config(state="normal")

def apply_feedback_colors(selected_index):
    question = quiz_data[current_question]
    correct_index = ord(question["answer"]) - 65

    for i, button in enumerate(choice_btns):
        button.config(state="disabled")
        if i == correct_index:
            button.config(style="success.TButton")
        elif i == selected_index:
            button.config(style="danger.TButton")
        else:
            button.config(style="secondary.TButton")

def next_question():
    global current_question
    if current_question < len(quiz_data) - 1:
        current_question += 1
        show_question()
    else:
        messagebox.showinfo(
            "Quiz gjennomført",
            f"Du fikk {score} av {len(quiz_data)} riktige.\nTakk for at du deltok!"
        )
        retry_btn.config(state="normal")
        next_btn.config(state="disabled")
        prev_btn.config(state="disabled")

def previous_question():
    global current_question
    if current_question > 0:
        current_question -= 1
        show_question()

# Start første gang
restart_quiz()
root.mainloop()
```
